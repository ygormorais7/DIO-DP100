```python
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv("aeroportoGuarulhos_2025.csv", sep=';')
```

```python
base.head(30)
```

```python
# 1. Fazemos uma cópia para não alterar a base original
df_tratado = base.copy()

# 2. Agrupamos por 'Data' e usamos ffill (forward fill) e bfill (backward fill).
# Isso faz com que os valores não-nulos (como Temp. Max e Chuva) preencham 
# os NaNs das outras horas do mesmo dia.
df_tratado = df_tratado.groupby('Data').transform(lambda x: x.ffill().bfill())

# 3. O comando .transform() remove a coluna que usamos para agrupar ('Data').
# Então, nós a devolvemos para o dataframe:
df_tratado['Data'] = base['Data']

# 4. Agora que os dados estão completos em todas as horas, filtramos só a linha das 12:00
# (Verifique se a sua coluna Hora é um número inteiro 1200 ou string "1200")
base_final = df_tratado[df_tratado['Hora (UTC)'] == 1200].copy()

# 5. Organizando o índice para ficar limpo (0, 1, 2, 3...)
base_final = base_final.reset_index(drop=True)

coluna_data = base_final.pop('Data')

# Insere a coluna 'Data' na posição 0 (primeira da esquerda)
base_final.insert(0, 'Data', coluna_data)

# Visualizando o resultado
display(base_final.head())
```

```python
base_final = base_final.drop(columns=['Hora (UTC)', 'Insolacao (h)'])
```

```python
# Verifica os tipos e valores das colunas
print("Tipos de dados:")
print(base_final.dtypes)
print("\nPrimeiras linhas:")
print(base_final.head())
print("\nNomes das colunas:")
print(base_final.columns.tolist())
```

```python
df = base_final.copy()

# Substitui vírgulas por pontos em TODAS as colunas numéricas
colunas_numericas = [
    "Temp. [Hora] (C)", 
    "Chuva [Diaria] (mm)", 
    "Umi. (%)", 
    "Nebulosidade (Decimos)",
    "Pressao (hPa)",
    "Vel. Vento (m/s)",
    "Dir. Vento (m/s)",
    "Temp. Max. [Diaria] (h)",
    "Temp. Min. [Diaria] (h)"
]

for col in colunas_numericas:
    df[col] = df[col].astype(str).str.replace(',', '.')
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Preenche NaN com a mediana (ou 0 para chuva)
df["Temp. [Hora] (C)"] = df["Temp. [Hora] (C)"].fillna(df["Temp. [Hora] (C)"].median())
df["Chuva [Diaria] (mm)"] = df["Chuva [Diaria] (mm)"].fillna(0)
df["Umi. (%)"] = df["Umi. (%)"].fillna(df["Umi. (%)"].median())
df["Nebulosidade (Decimos)"] = df["Nebulosidade (Decimos)"].fillna(df["Nebulosidade (Decimos)"].median())
df["Pressao (hPa)"] = df["Pressao (hPa)"].fillna(df["Pressao (hPa)"].median())
df["Vel. Vento (m/s)"] = df["Vel. Vento (m/s)"].fillna(df["Vel. Vento (m/s)"].median())
df["Dir. Vento (m/s)"] = df["Dir. Vento (m/s)"].fillna(df["Dir. Vento (m/s)"].median())
df["Temp. Max. [Diaria] (h)"] = df["Temp. Max. [Diaria] (h)"].fillna(df["Temp. Max. [Diaria] (h)"].median())
df["Temp. Min. [Diaria] (h)"] = df["Temp. Min. [Diaria] (h)"].fillna(df["Temp. Min. [Diaria] (h)"].median())

# padroniza data
df["Data"] = pd.to_datetime(df["Data"], dayfirst=True)

# cria variáveis de calendário
df["dia_semana"] = df["Data"].dt.dayofweek
df["fim_de_semana"] = (df["dia_semana"] >= 5).astype(int)
df["mes"] = df["Data"].dt.month

# feriados simples
feriados = [
    "2025-01-01", "2025-04-21", "2025-05-01",
    "2025-09-07", "2025-10-12", "2025-11-02",
    "2025-11-15", "2025-11-20", "2025-12-25"
]
df["feriado"] = df["Data"].astype(str).isin(feriados).astype(int)

np.random.seed(42)

# promoção mais comum em fim de semana
df["promocao"] = np.where(
    df["fim_de_semana"] == 1,
    np.random.binomial(1, 0.25, len(df)),
    np.random.binomial(1, 0.10, len(df))
)

# use os nomes REAIS das colunas
df["dia_quente"] = (df["Temp. [Hora] (C)"] >= 30).astype(int)
df["chuva_forte"] = (df["Chuva [Diaria] (mm)"] >= 10).astype(int)

ruido = np.random.normal(0, 8, len(df))

df["vendas_sorvete"] = (
    80
    + 7.0 * df["Temp. [Hora] (C)"]
    - 1.2 * df["Chuva [Diaria] (mm)"]
    - 0.15 * df["Umi. (%)"]
    - 0.8 * df["Nebulosidade (Decimos)"]
    + 22 * df["fim_de_semana"]
    + 28 * df["feriado"]
    + 18 * df["promocao"]
    + 12 * df["dia_quente"]
    - 10 * df["chuva_forte"]
    + ruido
)

# Remove qualquer NaN ou infinito
df["vendas_sorvete"] = df["vendas_sorvete"].fillna(80)
df["vendas_sorvete"] = df["vendas_sorvete"].replace([np.inf, -np.inf], 80)

df["vendas_sorvete"] = df["vendas_sorvete"].clip(lower=40).round().astype(int)

df.head()
```

```python
# Salva o dataframe em CSV com separador ";" e ponto como decimal
df.to_csv('vendas_sorvete_final.csv', sep=';', index=False, encoding='utf-8', decimal='.')
print("Arquivo 'vendas_sorvete_final.csv' salvo com sucesso!")
```
